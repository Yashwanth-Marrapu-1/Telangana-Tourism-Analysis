# List down the top 3 districts based on compounded annual growth rate (CAGR) of visitors between (2016 - 2019)?

CAGR Formula = (Vis_final/Vis_begin)^(1/t)

Vis_final = Final Value

Vis_begin = Begining Value

t= Time

```python
final_df_visitors.groupby(['district','year'])[['visitors']].sum()
```

```python
districts  = list(final_df_visitors['district'].unique())
```

```python
districts
```

```python
def CAGR(district):
    
    visitors = list(final_df_visitors[final_df_visitors['district'] == district].groupby('year')['visitors'].sum())
    
    Vis_begin = visitors[0]
    Vis_final = visitors[-1]
    t = 3
    
    if(Vis_begin!=0):
        
        CAGR = (Vis_final/Vis_begin)**(1/t)
        return district,(CAGR-1)*100
    else:
        return district,0
```

```python
CAGR_vis = pd.DataFrame([CAGR(i) for i in districts],columns=['District',"CAGR"])
CAGR_vis
```

```python
Top_CAGR_dist = CAGR_vis.sort_values(by='CAGR', ascending=False).head(3)
Top_CAGR_dist
```

```python
Top_CAGR_dist.plot("District", "CAGR", kind="bar")
```

# 7. What will be the projected number of domestic and foreign tourists in Hyderabad in 2025 based on the growth rate from previous years?

final_df_foreign_visitors ---> dataframe with all foreign visitors

final_df_domestic_visitors ---> dataframe with all domestic visitors

Growth Rate = ((Current Year Visitors - Previous Year Visitors) / Previous Year Visitors) * 100

```python
final_df_domestic_visitors[final_df_domestic_visitors['district'] == 'Hyderabad']
```

```python
final_df_foreign_visitors[final_df_foreign_visitors['district'] == 'Hyderabad']
```

```python
Hyderabad_domestic_visitors = final_df_domestic_visitors[final_df_domestic_visitors['district'] == 'Hyderabad'].groupby('year')['visitors'].sum()
```

```python
Hyderabad_domestic_visitors
```

```python
Hyderabad_foreign_visitors = final_df_foreign_visitors[final_df_foreign_visitors['district'] == 'Hyderabad'].groupby('year')['visitors'].sum()
```

```python
Hyderabad_foreign_visitors
```

```python
def PredVisitors(visitors,target_year) :
    Years = []
    predicted_visitors = []
    TotalGrowthRate = 0 # Total GrowthRate from 2016 to 2019
    
    for i in range(1,len(visitors)):
        TotalGrowthRate = TotalGrowthRate + ((visitors.values[i] - visitors.values[i-1])/visitors.values[i-1])*100
    
    Avg_GrowthRate = TotalGrowthRate / 3
    Pred_Visitors = visitors.values[-1] 
    diff_yrs = target_year - visitors.index[-1]
    
    for i in range(diff_yrs):
        Pred_Visitors = Pred_Visitors + (Pred_Visitors*Avg_GrowthRate)/100
        predicted_visitors.append(round(Pred_Visitors,ndigits=2))
        Years.append((visitors.index[-1])+(i+1))
    pd.DataFrame({"year":Years,"Aprox Visitors":predicted_visitors})
    
    return Years,predicted_visitors
```

```python
# Expected Hyderabad Foreign Visitors
Years, Foreign_predicted_visitors = PredVisitors(Hyderabad_foreign_visitors,2025)

pd.DataFrame({"Year":Years,"Predicted Visitors":Foreign_predicted_visitors})
```

```python
# Expected Hyderabad Domestic Visitors
Years, Domestic_predicted_visitors = PredVisitors(Hyderabad_domestic_visitors,2025)

pd.DataFrame({"Year":Years,"Predicted Visitors":Domestic_predicted_visitors})
```
